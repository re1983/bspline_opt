```python
import numpy as np

def plane_from_points(p1, p2):
    normal = np.cross(p1, p2)
    d = -np.dot(normal, p1)
    return normal, d

p1 = np.array([1, 2, 3])
p2 = np.array([4, 5, 6])
p1 = np.array([0, 0, 0])
p2 = np.array([10, 10, 10])

normal, d = plane_from_points(p1, p2)

print(f"Normal: {normal}")
print(f"D: {d}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_plane(p1, p2):
    normal = np.cross(p1, p2)
    d = -np.dot(normal, p1)
    xx, yy = np.meshgrid(range(10), range(10))
    zz = (-normal[0] * xx - normal[1] * yy - d) * 1. / normal[2]
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(xx, yy, zz)
    ax.scatter(p1[0], p1[1], p1[2], color='r')
    ax.scatter(p2[0], p2[1], p2[2], color='g')
    plt.show()

p1 = np.array([0, 0, 0])
p2 = np.array([10, 10, 10])
p1 = np.array([1, 2, 3])
p2 = np.array([4, 5, 6])

plot_plane(p1, p2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def project_onto_vector(point, vector):
    return point - np.dot(point, vector) * vector / np.dot(vector, vector)

p1 = np.array([0, 0, 0])
p2 = np.array([10, 10, 10])
v = p2 - p1

xx, yy = np.meshgrid(range(10), range(10))
zz = np.zeros_like(xx)

points = np.array([xx.ravel(), yy.ravel(), zz.ravel()]).T
projected_points = np.apply_along_axis(project_onto_vector, 1, points, v)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(projected_points[:, 0], projected_points[:, 1])
plt.show()
```
